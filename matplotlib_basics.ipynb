```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=np.arange(1,5)
y=x**3
plt.plot([1,2,3,4],[1,4,9,16],"go",x,y,'r^')
plt.title("FIRST PLOT")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot([1,2,3,4],[2,4,9,16],"go")
plt.title("First Subplot")
plt.subplot(1,2,2)
plt.plot([1,2,3,4],[2,4,9,16],'r^')
plt.title("Second Subplot")
plt.suptitle("My Subplots")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot([1,2,3,4],[2,4,9,16],"go")
plt.title("First Subplot")
plt.subplot(2,1,2)
plt.plot([1,2,3,4],[2,4,9,16],'r^')
plt.title("Second Subplot")
plt.suptitle("My Subplots")
plt.show()
```

```python
x = np.arange(1,5)
y = x**3
fig,ax = plt.subplots(nrows=3,ncols=2,figsize=(10,5))
ax[0,0].plot([1,2,3,4],[1,4,9,16],'b>')
ax[2,1].plot(x,y,'mo')
ax[0,0].set_title("Oval")
ax[2,1].set_title("Square")
plt.show( )
```

```python
divisions = ["div-A","div-B","div-C","div-D","div-E"]
avg_marks = [70,85,60,90,75]
variance=[1,5,8,7,6]
plt.bar(divisions, avg_marks, yerr=variance, color='b')
plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

```python
divisions = ["A","B","C","D","E"]
marks = [10,20,30,40,50]
plt.barh(divisions, marks, color='magenta')
plt.title("Bar Graph Horizontal")
plt.ylabel("Divisions")
plt.xlabel("Marks")
plt.show()
```

```python
divisions = ["A","B","C","D","E"]
marks = [80,80,80,80,80]
boys_marks = [60,65,70,75,65]
index=np.arange(5)
width = 0.3
plt.bar(index, marks, width,color='green', label="Total Marks")
plt.bar(index+width, boys_marks, width, color='red', label="Boy's Marks")
plt.title("Horizontal Stacked Bar")
plt.xlabel("Division")
plt.ylabel("Marks")
plt.xticks(index+width/2 , divisions)
plt.legend(loc="best")
plt.show()
```

```python
divisions = ["A","B","C","D","E"]
boys_marks = [60,65,70,75,65]
girls_marks = [65,70,75,75,70]
index=np.arange(5)
width = 0.3
plt.bar(index,girls_marks,width,color="blue",label="Girl's marks", bottom = boys_marks)
plt.bar(index, boys_marks, width, color="yellow", label="Boy's Marks")
plt.title("Vetically Stacked Bar")
plt.xlabel("Divisons")
plt.ylabel("Marks")
plt.xticks(index, divisions)
plt.legend(loc="best")
plt.show()
```

```python
plot = ["A","B","C","D","E"]
share=[100,200,300,100,400]
explode = [0,0,0,0,0.1]
plt.pie(share, explode=explode, labels=plot, startangle=10)
plt.axis('Equal')
plt.legend(title='List of Firms')
plt.show()
```

```python
x = np.random.randn(1000)
plt.title("Histogram")
plt.xlabel("Datas")
plt.ylabel("Frequency")
plt.hist(x,20)
plt.show()
```

## Practice Question

```python
g_students=[51,72,87,90,45]
b_students=[43,65,79,39,83]
clas=[1,2,3,4,5]
total= [x + y for x, y in zip(g_students,b_students)]
total
```

```python
fig,ax = plt.subplots(nrows=5,ncols=2,figsize=(20,20))

#Line Plot
ax[0,0].plot(clas, g_students,'green',label="Girl Students")
ax[0,0].plot(clas, b_students, 'blue', label="Boy Students")
ax[0,0].set_title("Line")
ax[0,0].set_xlabel("Class")
ax[0,0].set_ylabel("No of students")
ax[0,0].legend(loc="best")

#Scatter Plot
ax[0,1].scatter(clas, total,color='r')
ax[0,1].set_title("Scatter")
ax[0,1].set_xlabel("Class")
ax[0,1].set_ylabel("No of students")

#Bar Plot
variance = np.random.randn(5)
ax[1,0].bar(clas, total,yerr=variance, color='y')
ax[1,0].set_title("Bar Total")
ax[1,0].set_xlabel("Class")
ax[1,0].set_ylabel("No of students")

#Bar Horizontal
ax[1,1].barh(clas,total,color='m')
ax[1,1].set_title("Bar Horizontal")
ax[1,1].set_ylabel("Class")
ax[1,1].set_xlabel("No of students")

#Histogram
ax[2,0].set_title("Histogram")
ax[2,0].set_xlabel("Class")
ax[2,0].set_ylabel("No of students")
ax[2,0].hist(total,3)

#Pie
ax[2,1].pie(total, explode=explode, labels= clas, startangle=15)
ax[2,1].axis('Equal')
ax[2,1].set_title("Pie")
ax[2,1].legend(title='Class')

#3DScatter
ax[3,0] = fig.add_subplot(5,2,7, projection='3d')
ax[3,0].scatter3D(g_students, b_students, clas)
ax[3,0].set_title("3D Scatter Plot")
ax[3,0].set_xlabel("Girl students")
ax[3,0].set_ylabel("boy students")
ax[3,0].set_zlabel("class")

#3dLine
ax[3,1] = fig.add_subplot(5,2,8, projection='3d')
ax[3,1].plot3D(g_students, b_students, clas)
ax[3,1].set_title("3D Line Plot")
ax[3,1].set_xlabel("Girl students")
ax[3,1].set_ylabel("boy students")
ax[3,1].set_zlabel("class")

#StackedBarh
index=np.arange(1,6)
width = 0.3
ax[4,0].barh(index,g_students,width,color="blue",label="Girl Students", left = b_students)
ax[4,0].barh(index, b_students, width, color="yellow", label="Boy Students")
ax[4,0].set_title("Stacked Bar")
ax[4,0].set_xlabel("No of Students")
ax[4,0].set_ylabel("Class")
ax[4,0].set_yticks(index, clas)
ax[4,0].legend(loc="best")

#JointBarh
index=np.arange(1,6)
width = 0.3
ax[4,1].barh(index,g_students,width,color="blue",label="Girl Students")
ax[4,1].barh(index+width, b_students, width, color="yellow", label="Boy Students")
ax[4,1].set_title("Stacked Bar")
ax[4,1].set_xlabel("No of Students")
ax[4,1].set_ylabel("Class")
ax[4,1].set_yticks(index+width/2, clas)
ax[4,1].legend(loc="best")

fig.subplots_adjust(hspace=0.5, wspace=0.2)
fig.suptitle("MATPLOTLIB BASIC PLOTS", fontsize=30)
plt.show()
```
